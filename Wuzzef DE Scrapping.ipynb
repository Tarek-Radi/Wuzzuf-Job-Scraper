```python
jobs = []
for page in range(15):   
    print(f"Scraping page {page+1}...")

    url = f"https://wuzzuf.net/search/jobs?a=spbg&q=data%20engineer&start={page}"
    
    response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
    soup = BeautifulSoup(response.text, "html.parser")

    #Select Every h2 and a inside it
    title_links = soup.select("h2 a")
    print(title_links)

    # h2_tags = soup.find_all("h2")

    # title_links = []
    # for h2 in h2_tags:
    #     a = h2.find("a")
    #     if a:
    #         title_links.append(a)

    for link in title_links:
        title = link.get_text(" ", strip=True)
        href = link.get("href", "")
        job_link = "https://wuzzuf.net" + href if href.startswith("/") else href   #----because it was Exist in web 

        
        block = link.find_parent()

        
        parent_text = block.parent.get_text("\n", strip=True) if block and block.parent else ""
        # print(parent_text)
        # # Output :
        # # Middle Data Engineer
        # # Vertex Technologies  -
        # # Cairo,
        # # Egypt
        # # 3 days ago
        # # Senior Data Engineer
        
        lines = [line.strip() for line in parent_text.split("\n")]
        # print(lines)
        # Output ['Middle Data Engineer', 'Vertex Technologies  -', 'Cairo,', 'Egypt', '3 days ago']

        company = "N/A"
        city =    "N/A"
        country = "N/A"
        posted =  "N/A"

        if len(lines) > 0:
            title = lines[0]

        if len(lines) > 1:
            company = lines[1]

        if len(lines) > 2:
            city = lines[2]

        if len(lines) > 3:
            country = lines[3]

        if len(lines) > 4:
            posted = lines[4]

        jobs.append({
            "title": title,
            "company": company,
            "city": city,
            "country": country,
            "posted": posted,
            "link": job_link
        })

# for job in jobs[:2]:
#     print("Title   :", job["title"])
#     print("Company :", job["company"])
#     print("City    :", job["city"])
#     print("Country :", job["country"])
#     print("Posted  :", job["posted"])
#     print("Link    :", job["link"])
#     print("-" * 80)
```
